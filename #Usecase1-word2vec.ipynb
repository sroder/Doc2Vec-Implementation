```python
# Read the data
# Read Excel with Pandas
import pandas as pd
from pandas import ExcelFile
df = pd.read_excel('laufzeit_vertarg.xlsx',sheet_name = 'Sheet3')
df.head()
```

```python
# # check the description of Dataframe, which we read in previous step from the excel sheet
df.describe()
```

```python
# reading the scripts as sentences from the file
sentences_x = []
max_rows = 998
for i in df['Beschreibung']:
    sentences_x.append(i)
sentences_x
```

```python
#``RegexpTokenizer`` splits a string into substrings using a regular expression
# token_s stores the individual words
# for now keeping the stopwords in the sentences
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+\(\"\".+\n')
token_s=[]
for i in range(max_rows):
    token_s += [tokenizer.tokenize(sentences_x[i])]
```

```python
# check out how token_s looks like now
token_s[0][1]
```

```python
# remove stop words: A stop word is a commonly used word (such as “the”, “a”, “an”, “in”
# fwf stands for fixed width formatted lines. # pd.read_fwf(file.txt)
# read a text file using pandas as dataframe 
# but the problem is that Germainc Umlaut are not captured
# at the end converting txt file to excel and then using excel
stop_words_df = pd.read_excel('stopwords_de.xlsx',sheet_name = 'Sheet1')
#stop_words_list = stop_words_df.tolist()
stop_words_list = stop_words_df.values.tolist()
```

```python
# Given a list of words, remove any that are
# in a list of stop words.

def removeStopwords(wordlist, stopwords):
    return [w for w in wordlist if w not in stopwords]
```

```python
filtered =[]
for i in range(max_rows):
    filtered = removeStopwords(token_s[i],stop_words_list)
    print (filtered)
```

```python
# assigning a list of labels for each skript, the labels are mimics of actual sentences
# need to test here if Test sentences can be put as labels
LabelList = []
for i in range(max_rows):
    LabelList.append(tokenizer.tokenize(df['Skript'][i]))
```

```python
len(LabelList)
```

```python
# import modules & set up logging
import gensim, logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# train word2vec on Beschreibung 
model = gensim.models.Word2Vec(token_s, min_count=1)
```

```python
model.wv.most_similar(token_s[0])
```

```python
token_s[0]
```

```python
#build vocab
#model.build_vocab(token_s)
```

```python
 # get all sentence vectors: summing up over the snetence
Beschreibung_sent_vec_sum = []
for i in range(max_rows):
    Beschreibung_sent_vec_sum.append(sum(model.wv[token_s[i]]))
```

```python
len(Beschreibung_sent_vec_sum)
#
```

```python
# train another model word2vec on Skript 
model_label = gensim.models.Word2Vec(LabelList, min_count=1)
```

```python
 # get all label vectors: summing up per label
Label_sent_vec_sum = []
for i in range(max_rows):
    Label_sent_vec_sum.append(sum(model_label.wv[LabelList[i]]))
```

```python
#distance between Beschreibung_vec and Label_vec
from sklearn.metrics.pairwise import cosine_similarity
distance = []
for i in range(max_rows):
    distance_ = cosine_similarity(Beschreibung_sent_vec_sum[i],Label_sent_vec_sum[i])
    print(distance_)
```

```python
# putting the columns in exsiting Dataframe
#
#df = df[:499]
df['Beschreibung_vec'] = Beschreibung_sent_vec_sum 
df['Label_vec'] = Label_sent_vec_sum 
```

```python
######## RUN ONLY ONE TIME #############
for i in range(max_rows):
    df['Beschreibung_vec'][i] = sum(df['Beschreibung_vec'][i])
    df['Label_vec'][i] = sum(df['Label_vec'][i])
```

```python
# check out the new dataframe
df
```

```python
# We want to treat this as a classification problem from here on
# we would use list_docvec as the feature to train the model
# and Beschreibung as the target/ Label
# A primary model to be used will be SGD: Stochastic Gradient  Descent
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

#choosing a classifier
model_classifier = SGDClassifier(max_iter = 1500)

#defining Features and Target
#Features_ = df['list_vectors_docs'][:38] # taking 60 rows for Training
#Target_ =  new_df['Skript'][:38] # taking 60 rows for Training


```

```python
# train_test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Beschreibung_vec'],df['Skript'], test_size=0.20, random_state=0)
```

```python
# training the model
X_train = X_train.values.reshape(-1, 1)
model_classifier.fit(X_train,y_train)
```

```python
#score
X_test = X_test.values.reshape(-1, 1)
model_classifier.score(X_test,y_test)
```
