```python
# Read the data
# Read Excel with Pandas
import pandas as pd
from pandas import ExcelFile
df = pd.read_excel('laufzeit_vertarg.xlsx',sheet_name = 'Sheet1')
```

```python
# # check the rows of Dataframe, which we read in previous step from the excel sheet
df.head()
```

```python
# reading the scripts as sentences from the file
#sentences_x = [df['Skript'][0],df['Skript'][1]]
sentences_x = []
for i in df['Skript']:
    sentences_x.append(i)
sentences_x
```

```python
#``RegexpTokenizer`` splits a string into substrings using a regular expression
# token_s stores the individual words
# for now keeping the stopwords in the sentences
from nltk import RegexpTokenizer
from nltk.corpus import stopwords
tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+\(\"\".')
token_s=[]
for i in range(19):
    token_s += [tokenizer.tokenize(sentences_x[i])]
```

```python
# check out how token_s looks like now
token_s
```

```python
# assigning a list of labels for each skript, the labels are mimics of actual sentences
# need to test here if Test sentences can be put as labels
LabelList = []
max_rows = 19
for i in range(max_rows):
    LabelList.append(df['Beschreibung'][i])
```

```python
LabelList
```

```python
# label the doc
from gensim.models.doc2vec import LabeledSentence, TaggedDocument 
max_rows = 19
tag_doc = []
for i in range(max_rows):
    #print (i)
    tag_doc.append(TaggedDocument(token_s[i],LabelList[i]))
```

```python
# check words and tags pair in LabeledSentence
type(tag_doc)
tag_doc
```

```python
# if there are more than one sentence in TaggedDocument [ls], Doc2Vec model is failing 
from gensim.models.doc2vec import Doc2Vec
# initialize a model
model = Doc2Vec(vector_size=300, window=10, workers=1, alpha=0.025, min_alpha=0.01)
```

```python
#build vocab
model.build_vocab(tag_doc)
```

```python
 # get all document vectors
max_ = 19
list_docvec = []
list_vectors_docs = []
for i in range(max_):
    docvec = model.docvecs[i]
    list_docvec.append(docvec)
    vectors_docs = model.docvecs.vectors_docs[i]
    list_vectors_docs.append(vectors_docs)
```

```python
list_vectors_docs[0]
```

```python
# create a dataframe with columns: Beschreibung, list_docvec , list_vectors_docs and the skript
#
df = df[:19] # dropping the rows after 19th
df['list_docvec'] = list_docvec
df['list_vectors_docs'] = list_vectors_docs
```

```python
# check out the new dataframe
df
```

```python
# We want to treat this as a classification problem from here on
# we would use list_docvec as the feature to train the model
# and Beschreibung as the target/ Label
# A primary model to be used will be SGD: Stochastic Gradient  Descent
from sklearn.linear_model import SGDClassifier
model_classifier = SGDClassifier(random_state = None)
#Features_ = df['Beschreibung'][:18]
Features_ = df['list_vectors_docs'][:18]
Target_ =  df['list_docvec'][:18]
model_classifier.fit(Features_,Target_)
```
